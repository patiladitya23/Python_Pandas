```python
avocados = pd.read_csv('avocado.csv')
avocados
```

```python
# accessing unique regions in the dataset

avocados['region'].unique()
```

```python
avocados_totalUS = avocados[(avocados['region'] == 'TotalUS')
                           & (avocados['Date'] > '2018-01-01')]
avocados_totalUS
```

```python
avocados_totalUS = avocados_totalUS.sort_values(['Date'])

avocados_totalUS
```

```python
# creating pivot table using pandas

avocados_pivot = avocados_totalUS.pivot(index = 'Date',
                                        columns = 'type',
                                        values = 'AveragePrice')
avocados_pivot
```

```python
#multi indexing // hierarchical index

avocados = pd.read_csv('avocado.csv',index_col = ['region','Date'])
avocados = avocados.sort_index()
avocados
```

```python
avocados = avocados[avocados['type'] == 'organic']
avocados
```

```python
#applying condition to the index

avocados = avocados.loc[(avocados.index.get_level_values('Date') > '2018-03-01')]
avocados
```

```python
avocados.index
```

```python
#stack operation

avocados_pivot
```

```python
avocados_stack = avocados_pivot.stack()

avocados_stack
```

```python
avocados_unstack = avocados_stack.unstack()
avocados_unstack
```

```python
#melt function(similar to stacking a dataset)

avocados_melted = avocados.melt(id_vars = ['region','type','Date'],
                                value_vars = ['AveragePrice','Total Volume'])
avocados_melted
```
